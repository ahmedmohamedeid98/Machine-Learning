```python
# import Network
from Hand_written import HandWrittinRecognition

# import GUI 
from Gui import Ui_Form

# GUI lib needed
from PyQt5 import QtCore, QtGui, QtWidgets
from PyQt5.QtWidgets import QWidget, QApplication
from PyQt5.QtCore import *
from PyQt5.QtGui import *
import imageio
import sys
import numpy
import re
import matplotlib.pyplot as plt
import matplotlib
matplotlib.use('Agg')

input_nodes = 784
hidden_nodes = 200
output_nodes = 10
learning_rate = 0.01

# create instance of neural network
Network = HandWrittinRecognition(input_nodes,hidden_nodes,output_nodes, learning_rate)

#Network.train_model()

###############################################################################
#                                GUI Main Function                            #
###############################################################################
class main(QWidget, Ui_Form):
    def __init__(self):
        QWidget.__init__(self)
        self.setupUi(self)
        
        self.pushButton.clicked.connect(self.Browse)
        self.pushButton_2.clicked.connect(self.Recognize)
        self.pushButton_3.clicked.connect(self.Train_Network)
        self.show_info.clicked.connect(self.Show_info)
    def Browse(self):
      
        global fileName
        
        fileName, _ = QtWidgets.QFileDialog.getOpenFileName(self, 'Single File','my_own_images' , '*.png')#QtCore.QDir.rootPath()
        new_path = fileName
        
        self.label_3.setPixmap(QPixmap(re.sub(r'my_own_images','NOrginal',new_path)))
        self.pushButton_2.setEnabled(True)
        
    
    def Recognize(self):
        img_path = fileName
        img_array = imageio.imread(img_path,as_gray=True)#, as_gray=True
        img_data  = 255.0 - img_array.reshape(784)
        img_data = (img_data / 255.0 * 0.99) + 0.01
        network_output = Network.get_output(img_data)
        x = numpy.arange(0,10)
        y = numpy.array(network_output).reshape(10,)
        plt.bar(x,height=y,width=0.5,tick_label=x)
        plt.xlabel("Digites")
        plt.ylabel("network output")
        plt.savefig("dynamic_images/Digite_weight",dpi=50)
        plt.close()
        self.Digites_plot.setPixmap(QPixmap('dynamic_images/Digite_weight.png'))
        self.label.setText("Network Say: "+str(numpy.argmax(network_output)))
       
    def Train_Network(self):
        num_times_use_dataset = 10
        acc = ""
        x = numpy.arange(1,num_times_use_dataset+1)
        y = []
        for i in range(num_times_use_dataset):
            acc = Network.train_model()
            self.accuarcy_label.setText(acc)
            self.progressBar.setProperty("value", 10*(i+1))
        for line in acc.split('\n')[:-1]:
            y.append(float(line))
            
            
        plt.plot(x,y)
        plt.legend('accuracy line')
        plt.xlabel("Rounds")
        plt.ylabel("Accuracy")
        plt.savefig("dynamic_images/ploting",dpi=55)
        plt.close()
        self.pushButton.setEnabled(True)
        self.ploting_label.setPixmap(QPixmap('dynamic_images/ploting.png'))
    
    def Show_info(self):
        info = 'Used Algorithm: Backpropogation'
        info += '\n'
        info += 'Transfer Function: Sigmod'
        info += '\n'
        info += 'Number of Layers : 2'
        info += '\n'
        info += 'Input Vector: [784 x 1]'
        info += '\n'
        info += 'Input Weight: [200 x 784]'
        info += '\n'
        info += 'Output Weight: [10 x 200]'
        
        self.label_5.setText(info)
    

        
app = QApplication(sys.argv)
window = main()
window.show()
app.exec_()
```
